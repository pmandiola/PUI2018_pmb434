```python
from __future__ import print_function
__author__= 'fbb'

import numpy as np
import matplotlib.pylab as plt
import seaborn
from scipy.optimize import curve_fit, minimize
%pylab inline
```

```python
#Set the seed
np.random.seed(2018)
```

```python
#Set the common mean for all distributions
common_mean = 100
```

```python
#Create empty dictionary to store our distributions
distributions = {}
```

### Chi-Squared distribution

```python
distributions['chisq'] = {}

#Generate 100 sample sizes
sample_sizes = np.random.randint(low=11, high=2000, size=100)

for n in sample_sizes:
    #One dictionary for each size
    distributions['chisq'][n] = {}
    
    #Save the sample size (it's also the key)
    distributions['chisq'][n]['size'] = n
    
    #Generate the sample
    distributions['chisq'][n]['sample'] = np.random.chisquare(df=common_mean, size = n)
    
    #Save the mean
    distributions['chisq'][n]['mean'] = distributions['chisq'][n]['sample'].mean()
```

```python
ax = plt.figure(figsize=(10,6)).add_subplot(111)

ax.set_xlabel('Sample size', fontsize=18)
ax.set_ylabel('Sample mean', fontsize=18)
ax.set_title('Figure 1: Chi Squared', fontsize=18)
ax.plot([min(sample_sizes), max(sample_sizes)], [common_mean, common_mean], 'k')

for n in distributions['chisq']:
    #Plot each sample mean
    ax.plot(n, distributions['chisq'][n]['mean'], 'o')
```

Figure 1: With the Chi Squared distribution we can see that as the sample size increases, the sample mean is closer to the mean of the distribution. This is consistent with the law of large numbers, which states that the average of the results obtained from a large number of trials should be close to the expected value, and will tend to become closer as more trials are performed.

### Binomial distribution

```python
distributions['binomial'] = {}

#Generate 100 sample sizes
sample_sizes = np.random.randint(low=11, high=2000, size=100)

for n in sample_sizes:
    #One dictionary for each size
    distributions['binomial'][n] = {}
    
    #Save the sample size (it's also the key)
    distributions['binomial'][n]['size'] = n
    
    #Generate the sample (need n*p = common_mean)
    distributions['binomial'][n]['sample'] = np.random.binomial(n = 200, p = common_mean/200, size = n)
    
    #Save the mean
    distributions['binomial'][n]['mean'] = distributions['binomial'][n]['sample'].mean()
```

```python
ax = plt.figure(figsize=(10,6)).add_subplot(111)

ax.set_xlabel('Sample size', fontsize=18)
ax.set_ylabel('Sample mean', fontsize=18)
ax.set_title('Figure 4: Binomial', fontsize=18)
ax.plot([min(sample_sizes), max(sample_sizes)], [common_mean, common_mean], 'k')

for n in distributions['binomial']:
    #Plot each sample mean
    ax.plot(n, distributions['binomial'][n]['mean'], 'o')
```

Figure 4: Again, with the Binomial distribution we can see that as the sample size increases, the sample mean is closer to the mean of the distribution. This is consistent with the law of large numbers.

### Geometric distribution

```python
distributions['geometric'] = {}

#Generate 100 sample sizes
sample_sizes = np.random.randint(low=11, high=2000, size=100)

for n in sample_sizes:
    #One dictionary for each size
    distributions['geometric'][n] = {}
    
    #Save the sample size (it's also the key)
    distributions['geometric'][n]['size'] = n
    
    #Generate the sample (need n*p = common_mean)
    distributions['geometric'][n]['sample'] = np.random.geometric(p = 1/common_mean, size = n)
    
    #Save the mean
    distributions['geometric'][n]['mean'] = distributions['geometric'][n]['sample'].mean()
```

Figure 5: Again, with the Geometric distribution we can see that as the sample size increases, the sample mean is closer to the mean of the distribution. This is consistent with the law of large numbers.

### Plotting the distribution of all sample means together

```python
#Create a single list with all the sample means
all_sample_means = []

for dist in distributions.values():
    for n in dist.values():
        all_sample_means.append(n['mean'])
```

```python
ax = plt.figure(figsize=(10,6)).add_subplot(111)

#ax.set_xlabel('Sample size', fontsize=18)
#ax.set_ylabel('Sample mean', fontsize=18)
ax.set_title('Figure 6: Histogram of all sample means', fontsize=18)
ax.hist(all_sample_means, bins=100)
plt.show()
```

Figure 6: As the Central Limit Theorem establishes, in this histogram we can see that the means of all our samples (independently of the distribution) tend toward a normal distribution around the common mean we set. In our case, we have a very narrow normal distribution with mean 100.
